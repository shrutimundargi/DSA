## 1. Find if target is sum of two elements in an array, given array with n elements, target k is sum of any two elements.

### Naive approach

```python
arr = [2, 3, 4, 5, 6, 7, 8]
target = 7
arr2 = []

for i in arr:
  for j in arr:
    if i+j == target:
      if [j,i] not in arr2:
        arr2.append([i,j])

print(arr2)
```

## Two-pointer approach

We will first sort our array. Once our array is sorted, the sum of two elements is going to be either smaller or bigger than our target number.


If the target is ***bigger*** we need to ***reduce*** the sum of elements so we will ***decrement*** our right pointer.

If the sum is ***smaller*** we need to ***increase*** the sum of elements and hence we will ***increment*** our left pointer.

```python
arr = [15,26,45,78,90]
arr.sort()
target = 135
arr2 = []
first = 0
second = len(arr)-1
print("Initial assignment:",first,second)

while first < second:
  if arr[first] + arr[second] == target:
    arr2.append([arr[first],arr[second]])
    first+=1
    second = len(arr)-1
  elif arr[first]+ arr[second] > target:
    print("sum is bigger",arr[first], arr[second])
    second-=1
  else:
    print("sum is smaller",arr[first], arr[second])
    first+=1

print(arr2)
```

## 2. Check if the string is palindrome.

```python
pal = "banana"
left, right = 0, len(pal)-1
while left < right:
  if pal[left] == pal[right]:
    left+=1
    right-=1
  else:
    print("Not a palindrome")
    break
  print("is a palindrome")
```
